## 17. Найдите процентное соотношение учащихся, сдавших экзамен на «отлично», «хорошо», «удовлетворительно», «неудовлетворительно», задав самостоятельно соответствующие критерии и учтя минимальный балл.

```python
mark_5 = df['Балл'][(df['Балл'] > 72)&(df['Балл']<=100)].count() *100/df.shape[0]
mark_4 = df['Балл'][(df['Балл'] > 24)&(df['Балл']<=57)].count() *100/df.shape[0]
mark_3 = df['Балл'][(df['Балл'] > 57)&(df['Балл']<=72)].count() *100/df.shape[0]
mark_2 = df['Балл'][df['Балл'] < df['Минимальный балл']].count() *100/df.shape[0]
print(mark_2,mark_3,mark_4,mark_5)
```

## 18. Постройте круговую диаграмму, показывающую распределение учащихся, сдавших экзамен на «отлично», «хорошо», «удовлетворительно», «неудовлетворительно».

```python
lab =['отлично', 'хорошо', 'удовлетворительно', 'неудовлетворительно']
values = [mark_5,mark_4,mark_3,mark_2]
plt.pie(values, labels = lab, autopct = '%5.1f%%', explode = [0.05, 0.05, 0.05, 0], radius = 1.5, startangle= 90)
plt.title('распределение учащихся, сдавших экзамен на «отлично», «хорошо», «удовлетворительно», «неудовлетворительно»')
plt.show()
```
