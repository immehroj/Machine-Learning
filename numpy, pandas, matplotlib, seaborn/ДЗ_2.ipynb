Постройте гистограмму по целевой переменной Price. Оцените визуально, по какой цене продаётся наибольшее количество квартир.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
df
```

```python
import matplotlib.pyplot as plt

# Получите данные цены квартир
prices = df['Price']

# Постройте гистограмму
plt.hist(prices, bins=20)

# Добавьте подписи к осям
plt.xlabel('Price')
plt.ylabel('Count')

# Отобразите гистограмму
plt.show()
```

```python
prices = df['Price']

# Постройте гистограмму с использованием seaborn
sns.histplot(prices, bins=20, color='red')

# Добавьте подписи к осям
plt.xlabel('Price')
plt.ylabel('Count')

plt.show()
```

Постройте диаграммы рассеяния для признаков Rooms, Square, HouseFloor, HouseYear в зависимости от целевой переменной Price в одной области figure. Оцените визуально, есть ли среди них такие, на которых разброс точек близок к линейной функции.

```python
f = plt.figure(figsize = (16,10))
ax = f.subplots(2,2)
ax[0,0].scatter(df['Rooms'], df['Price'])
ax[0,1].scatter(df['Square'], df['Price'])
ax[1,0].scatter(df['HouseFloor'], df['Price'])
ax[1,1].scatter(df['HouseYear'], df['Price'])

plt.suptitle('диаграммы рассеяния для признаков Rooms, Square, HouseFloor, HouseYear в зависимости от целевой переменной Price в одной области figure.', fontsize = 20)
plt.show()
```

Постройте ядерную оценку плотности целевой переменной Price. Оцените визуально, напоминает ли полученный график нормальное распределение. Постройте двумерную ядерную оценку плотности для целевой переменной Price и признака HouseFloor, затем оцените визуально на каких этажах и по какой цене продаётся основная масса квартир.

```python
#оценка плотности целевой переменной Price
sns.kdeplot(df['Price'], shade=True)
plt.xlabel('Price')
plt.ylabel('Density')
plt.title('Kernel Density Estimation of Price')
plt.show()
```

```python
# Построение двумерной ядерной оценки плотности для целевой переменной Price и признака HouseFloor
sns.kdeplot(df['Price'], df['HouseFloor'], fill=True, cmap='Blues')
plt.xlabel('Price')
plt.ylabel('HouseFloor')
plt.title('Kernel Density Estimation of Price and HouseFloor')

plt.show()
```
